```python
funcsites_19 = pd.read_csv('/Users/asjaeger/Downloads/functional_sites_2019_bygene/output.tsv', sep='\t')
funcsites_19
```

```python
funcsites_19 = funcsites_19.dropna(subset=['Sequence Feature Name'])
funcsites_19
```

```python
funcsites_19['Feature Identifier'] = funcsites_19['Feature Identifier'].str.replace(r'^Influenza A_', '', regex=True)
```

```python
funcsites_19['gene'] = funcsites_19['Feature Identifier'].str.extract(r'^([^_]+)')
funcsites_19
```

```python
funcsites_19['gene'].unique()
```

```python
import numpy as np

# Define mapping: gene(s) -> segment
segment_map = {
    "HA": ["H5"],
    "NA": ["N1",],
    "NEP": ["NS2"],
    "NP": ["NP"],
    "NS1": ["NS1"],   
    "PA": ["PA"],
    "PB1": ["PB1"],
    "PB2": ["PB2"]
}

# Build conditions and choices for np.select
conditions = []
choices = []

for segment, genes in segment_map.items():
    conditions.append(funcsites_19['gene'].isin(genes))
    choices.append(segment)

# Default value if no match found
default_segment = "Unknown"

# Apply logic
funcsites_19['segment'] = np.select(conditions, choices, default=default_segment)
funcsites_19
```

```python
funcsites_1aa = funcsites_19[funcsites_19['Sequence Feature Length'] == 1]
funcsites_1aa
```

```python
funcsites_1aa[['change_info', 'Amino_Acid']] = funcsites_1aa['Sequence Feature Name'].str.extract(
    r'Influenza A_[^_]+_([^-]+(?:-[^_]+)*)_(\d+)'
)
funcsites_1aa
```

```python
funcsites_1aa.columns
```

```python
funcsites_1aa.drop(columns=['Feature Identifier', 'Sequence Feature Name',
       'Sequence Feature Length','Amino Acid Position', 'Evidence', 'Comments', 'Unnamed: 8'], inplace=True)
funcsites_1aa
```

```python
funcsites_1aa = funcsites_1aa.rename(columns={'Category': 'Type', 'segment': 'Gene'})
funcsites_1aa.drop_duplicates(subset='Amino_Acid', keep='first', inplace=True, ignore_index=False)
funcsites_1aa
```
