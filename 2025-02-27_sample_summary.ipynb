```python
import pandas as pd
```

```python
#reading in list of good samples by sample/rep/segment
good = pd.read_csv('good_consesnsus.tsv', sep = '\t')
good
```

```python
good.columns
```

```python
good = good.drop(['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6'], axis=1)
good
```

```python
# good[['sample', 'rep', 'segment']] = good['sample_rep_segment'].str.split('_', expand=True)
# good
good[['sample', 'rep', 'segment']] = good['sample_rep_segment'].str.rsplit('_', n=2, expand=True)

good
```

```python
good.to_csv('good_consensus_parsed.tsv', sep='\t', index=False)
```

```python
# Value counts for the 'segment' column
segment_counts = good['segment'].value_counts()

# Value counts for the 'gene' column
sample_counts = good['sample'].value_counts()

# Display the results
print("Segment counts:")
print(segment_counts)

print("\nSample counts:")
print(sample_counts)
```
